```python
import os
import json
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

file = open (os.path.relpath('../test/log.json', os.path.abspath('')))
data = json.load(file)

figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')

num = np.array([e["n"] for e in data])
mean = np.array([np.mean(e["t"]) for e in data])
std = np.array([np.std(e["t"], ddof=1) for e in data])
nlogn = np.array([0.03*e["n"]*math.log(e["n"], 2) for e in data])

plt.plot(num, mean, color='green', label="Mdia")
plt.plot(num, std, color='blue', label="Desvio padro")
plt.plot(num, nlogn, color='red', label="O(nlog(n))")

plt.ylabel('Tempo de execuo (microssegundos)',fontsize=14)
plt.xlabel('Nmero de ilhas',fontsize=14)

plt.legend()
plt.show()
```
