```python
import os
import json
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

file_pd = open (os.path.relpath('../test/log_pd.json', os.path.abspath('')))
file_greedy = open (os.path.relpath('../test/log_greedy.json', os.path.abspath('')))

data_pd = json.load(file_pd)
data_greedy = json.load(file_greedy)

num_pd = np.unique([e["n"] for e in data_pd])
num_greedy = np.unique([e["n"] for e in data_greedy])
```

```python
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')

mean = np.array([np.mean(data_pd[i*3]["t"]) for i in range(len(data_pd)//3)])
mean2 = np.array([np.mean(data_pd[i*3 + 1]["t"]) for i in range(len(data_pd)//3)])
mean3 = np.array([np.mean(data_pd[i*3 + 2]["t"]) for i in range(len(data_pd)//3)])

plt.title("Programao Dinmica - Mdia", fontsize=16)

plt.plot(num_pd, mean, color='green', linewidth=3, markersize=10, label="Mdia para N=100xM")
plt.plot(num_pd, mean2, color='blue', linewidth=3, markersize=10, label="Mdia para N=100xM/2")
plt.plot(num_pd, mean3, color='red', linewidth=3, markersize=10, label="Mdia para N=100xM/3")

plt.ylabel('Tempo de execuo (microssegundos)',fontsize=14)
plt.xlabel('Nmero de ilhas',fontsize=14)

plt.legend()
plt.show()
```

```python
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')

std = np.array([np.std(data_pd[i*3]["t"], ddof=1) for i in range(len(data_pd)//3)])
std2 = np.array([np.std(data_pd[i*3 + 1]["t"], ddof=1) for i in range(len(data_pd)//3)])
std3 = np.array([np.std(data_pd[i*3 + 2]["t"], ddof=1) for i in range(len(data_pd)//3)])

plt.plot(num_pd, std, color='green', linewidth=3, markersize=10, label="Desvio padro M=100xN")
plt.plot(num_pd, std2, color='blue', linewidth=3, markersize=10, label="Desvio padro M=100xN/2")
plt.plot(num_pd, std3, color='red', linewidth=3, markersize=10, label="Desvio padro M=100xN/3")

plt.ylabel('Tempo de execuo (microssegundos)',fontsize=14)
plt.xlabel('Nmero de ilhas',fontsize=14)

plt.legend()
plt.show()
```

```python
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')

num = np.array([e["n"] for e in data_greedy])
mean = np.array([np.mean(e["t"]) for e in data_greedy])
std = np.array([np.std(e["t"], ddof=1) for e in data_greedy])
nlogn = np.array([0.03*e["n"]*math.log(e["n"], 2) for e in data_greedy])

plt.title("Guloso", fontsize=16)

plt.plot(num, mean, color='green', label="Mdia")
plt.plot(num, std, color='blue', label="Desvio padro")
plt.plot(num, nlogn, color='red', label="O(nlog(n))")

plt.ylabel('Tempo de execuo (microssegundos)',fontsize=14)
plt.xlabel('Nmero de ilhas',fontsize=14)

plt.legend()
plt.show()
```
